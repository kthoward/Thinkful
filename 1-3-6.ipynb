```python
'''Making a dataframe and cleaning up the column names'''
df = pd.read_csv('wellcome_apcspend2013_forthinkful.csv')
df.columns = ['PMID/PMCID', 'publisher', 'journal_title', 'article_title', 'cost']
print(df)
```

```python
'''Removing anything but the numbers in the cost column'''
cost = df['cost']
cost_2 = cost.apply(lambda x: ''.join(list(filter(str.isdigit, str(x)))))
df['cost'] = cost_2
print(df)
```

```python
'''Getting rid of null values'''
df = df.dropna()
print(df)
```

```python
'''Making all of the journals lowercase so we can aggregate them'''
journals = df['journal_title']
journals = journals.str.lower()
journals = journals.str.strip()
df['journal_title'] = journals
print(df)
```

```python
'''Removing extra characters and spaces'''
journal = df['journal_title']
jt = journal.apply(lambda x: ''.join(list(filter(str.isalpha, str(x)))))
df['journal_title'] = jt
print(df['journal_title'])
```

```python
'''Grouping by journal_title'''
y = df.groupby('journal_title').count()
print(y)
```

```python
'''Sorting in descending order to find the five most common journals and the number of articles associated with themthey are:
-plos_one: 197
-journal_of_biological_chemistry: 52
-neuro_image: 28
-nucleic_acid_research: 25
-plos_pathogens: 24'''
y.sort_values(by='PMID/PMCID', ascending=0)
```

```python
change = df['cost']
df['cost'] = pd.to_numeric(change)
a = df.groupby("journal_title")[['cost']].mean()
print(a)
    
```

```python
print(df.groupby("journal_title")[['cost']].median())
```
