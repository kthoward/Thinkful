```python
features = pd.get_dummies(df['pos'])

features['wings'] = np.where((df['pos'].isin(['SG', 'SF', 'PF'])), 1, 0)

print(pd.crosstab(features['wings'], df['pos']))
```

```python
features['above_average'] = np.where(df['ppm']>=.808, 1, 0)

print(df['ppm'].groupby(features['above_average']).describe())
```

```python
features['role_player'] = np.where(df['min']<=20, 1, 0)

print(df['min'].groupby(features['role_player']).describe())
```

```python
features["reb_dk"] = df['reb'] * 1.25
features.reb_dk.head(10)
```

```python
features['points_dk'] = df['pts']
features.head(10)
```

```python
features['assist_dk'] = df['assist'] * 1.5
features.head(100)
```

```python
features['gunner'] = np.where(df['3pa']>=2.5, 1, 0)

print(df['ppm'].groupby(features['gunner']).describe())
```

```python
features['offensive_rebounder'] = np.where(df['oreb']>=1, 1, 0)

print(df['ppm'].groupby(features['offensive_rebounder']).describe())
```

```python
features['defender'] = np.where(df['bl']>=.4, 1, 0) | np.where(df['st']>=.81, 1, 0)

print(df['ppm'].groupby(features['defender']).describe())
```
