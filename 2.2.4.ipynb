```python
df = pd.read_csv('adult.data.csv', header = 0)
df.columns = ['age', 'Workclass', 'fnlwgt' , 'Education', 'Education_num', 'Marital_status', 'Occupation', 'Relationship', 'Race', 'Sex', 'Capital_gained', 'Capital_lost', 'Hours_per_week', 'Native_Country', 'Income']
df.head(10)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df.ix[1:100, 'age'], color='green')
plt.ylabel('Age')
plt.title('Age')

plt.subplot(1, 2, 2)
plt.plot(df.ix[1:100, 'Education_num'], color='purple')
plt.ylabel('Years')
plt.title('Years of Schooling')
# Here I am looking at subpolts of the distributions of a sample of ages as well as years of schooling.
# I want to see how large of a swing there is and where the data seems to centralize. It appears that the most frequent
# ages fall between 30 and 60 and the most frequent years of schooling falls between 9 and 13.
plt.show()
```
