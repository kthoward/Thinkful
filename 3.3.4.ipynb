```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lgstc = np.random.logistic(15, 1, 100)
plt.hist(lgstc)
plt.axvline(lgstc.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(lgstc.mean() + lgstc.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(lgstc.mean() - lgstc.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
# The majority of the datapoints are within one std of the mean with a steep dropoff beyond that. It is interesting 
# that the majority of the datapoints within one std of the mean are the same value
```

```python
l_lace = np.random.laplace(5, 1, 100)
plt.hist(l_lace)
plt.axvline(l_lace.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(l_lace.mean() + l_lace.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(l_lace.mean() - l_lace.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
# Here we see a large dropoff from the mean to the rest of the datapoints even though the majority of the points still 
# rest within one std of the mean. It is nteresting that so far two of the histograms share similarities with normal 
# distribution.
```

```python
ray = np.random.rayleigh(4, 100)
plt.hist(ray)
plt.axvline(ray.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(ray.mean() + ray.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(ray.mean() - ray.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
# This graph is definitely different then normal distribution. It seems like an instance of "push and pull" where the 
# initial force becomes weaker as if it were a person lifting a weight and becoming fatigued.
```

```python
vonm = np.random.vonmises(5, 10, 100)
plt.hist(vonm)
plt.axvline(vonm.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(vonm.mean() + vonm.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(vonm.mean() - vonm.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
# This seems to resemble normal distribution except on the right side of the graph. The majority of the datapoints are 
# between the mean and one std.
```

```python
logs = np.random.logseries(.4, 100)
plt.hist(logs)
plt.axvline(logs.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logs.mean() + logs.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(logs.mean() - logs.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
# The majority of the datapoints are within the mean and std of the mean but it resembles an exponential curve that 
# has been flipped. 
```

```python
wld = np.random.wald(5, 3, 100)
plt.hist(wld)
plt.axvline(wld.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(wld.mean() + wld.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(wld.mean() - wld.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
# This is similar to the graph above as it has the majority or the datapoints between one std of the mean and it 
# resembles a flipped exponential curve. It differs in the fact that it offers more diverse values but the dropoff
# seems similar. I would wonder if these have similarities in the natural worls.
```

```python
rand1 = np.random.normal(5, .5, 100)
rand2 = np.random.normal(10, 1, 100)
rand3 = np.random.normal(15, 1.5, 100)
plt.axvline(rand3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean() - rand3.std(), color='b', linestyle='dashed', linewidth=2) 
plt.hist(rand3)
# Added together this seems like a more clearly defined example of normal distribution then rand1 and rand 2 that I
# have graphed below for comparison
```
