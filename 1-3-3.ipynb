```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
os.getcwd()
os.chdir('/Users/kylehoward/desktop/thinkful/') 
os.getcwd()
```

```python
df = pd.DataFrame(pd.read_csv('pro_projections.csv'))
df.columns = ['player', 'position', 'id', 'draft_year', 'projected_spm', 'superstar', 'starter', 'role_player', 'bust']
print(df)
```

```python
w = df['starter']
x = df['projected_spm']
y = df['superstar']
z = df['draft_year']
''' This is a simple way to look at data distribution with the added benefit of the kdeplot to help visualize. It isn't
very valuable outside of viewing initial distribution '''
sns.distplot(x)
sns.plt.title('Plot 1: Histogram plot with kdeplot for projected spm')
```

```python
''' A boxplot is a great way to view outliers as well as distribution and relationship with the mean. The con is that
it is not very visually appealing. '''
sns.boxplot(x,orient='v')
sns.plt.title('Plot 2:  Inverted boxplot for projected spm')
```

```python
''' A violinplot is basically a combination of a histogram mixed with a boxplot. It is visually more appealing then a 
boxplot but can appear deceivingly smooth with small sample sizes'''
sns.violinplot(x)
sns.plt.title('Plot 3: Violin plot with for projected spm')
```

```python
''' A point polt can show the mean and relationship to the mean with multiple varaibles. It is pretty limited in this 
scope but performs the task of relativity better then a bar graph'''
sns.pointplot(x)
sns.plt.title('Plot 4: Line plot for projected spm')
```

```python
''' Plots the residual of linear regression. It is not very visually appealing and is difficult to read. '''
sns.residplot(w,y)
sns.plt.title('Plot 5: Scatter showing relationship between superstar and starter odds')
```

```python
'''I need to format this in a more visually appealing way.'''
sns.pointplot(w,y)
sns.plt.title('Plot 6: Pointplot showing relationship between superstar and starter odds')
```

```python
''' Swarmplot works best when used in conjunction with a violin plot or a boxplot. It does not scale up well'''
sns.swarmplot(w,y)
sns.plt.title('Plot 7: Swarmplot showing relationship between superstar and starter odds')
```

```python
'''This helps to show if there is structure to the residuals'''
sns.residplot(z,x)
sns.plt.title('Plot 8: Residplot showing relationship between draft year and projected spm')
```

```python
'''Plots data and linear regression model'''
sns.regplot(z,x)
sns.plt.title('Plot 9: Regplot showing relationship between draft year and projected spm')
```

```python
''' Draws a scatterplot where one variable is conditional '''
sns.stripplot(z,x)
sns.plt.title('Plot 10: Stripplot showing relationship between draft year and projected spm')
```

```python
'''Shows the mean and standard error. Can be deceiving and inefficient '''
sns.barplot(z,x)
sns.plt.title('Plot 11: barplot showing relationship between draft year and projected spm')
```

```python
''' My favorite plot. Shows outliers, distribution, and mean. '''
sns.violinplot(z,x)
sns.plt.title('Plot 12: Violinplot showing relationship between draft year and projected spm')
```
